```python
import matplotlib.pyplot as plt

# Filter rows where Origin is 'Asia'
asia_make = data[data['Origin'] == 'Asia']

# Count the occurrences of each unique 'Make'
make_counts = asia_make['Make'].value_counts()

# Plot a bar chart
plt.figure(figsize=(10, 6))
make_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Cars by Make (Origin: Asia)')
plt.xlabel('Car Make')
plt.ylabel('Number of Cars')
plt.show()
```

```python
# pie chart based on the 'MPG_Highway' values
```

```python
import matplotlib.pyplot as plt

# Sort the entire dataset by 'MPG_Highway' in ascending order
sorted_data = data.sort_values(by='MPG_Highway', ascending=True)

# Define bins for categorizing 'MPG_Highway'
bins = [0, 20, 30, 40, 50, float('inf')]  # Adjust the bin edges as needed

# Create labels for the bins
labels = ['0-20', '21-30', '31-40', '41-50', '51+']

# Create a new column 'MPG_Highway_Category' based on bins
sorted_data['MPG_Highway_Category'] = pd.cut(sorted_data['MPG_Highway'], bins=bins, labels=labels, right=False)

# Count the occurrences of each category
category_counts = sorted_data['MPG_Highway_Category'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of MPG_Highway Categories')
plt.show()
```
